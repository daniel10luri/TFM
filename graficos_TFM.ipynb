```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.rcParams['font.family'] = 'sans-serif'
```

```python
# Cargar el dataframe desde el archivo CSV
df = pd.read_csv("https://docs.google.com/spreadsheets/d/1vFv4F3e-X1IkBkcuUobK5snUWI2R_5Y7/export?format=csv")
# Suprimir la primera columna
df_genes = df.iloc[:, 1:]
# Cuenta cuántas veces un gen aparece (valor diferente a 0)
prevalencia_genes = (df_genes != 0).sum(axis=0)

# Ordenar los genes por prevalencia de menor a mayor
prevalencia_genes = prevalencia_genes.sort_values(ascending=False)

# Crear un gráfico de barras
plt.figure(figsize=(10, 6))
prevalencia_genes.plot(kind='bar', color='lightsalmon', width=0.7, legend=False)

# Configuración del gráfico
plt.title('Prevalencia de Genes en las Muestras')
plt.xlabel('Genes')
plt.ylabel('Prevalencia')
plt.xticks(rotation=90, fontsize=8)
plt.ylim(0, prevalencia_genes.max() * 1.2)
plt.tight_layout()  # Para ajustar el gráfico de manera que no se corte la etiqueta

# Mostrar el gráfico
plt.show()
```

```python
# Cargar los datos directamente desde la URL
df = pd.read_csv("https://docs.google.com/spreadsheets/d/1AVefl9QtHKURb3vFBHtjgDNojQPUTClm/export?format=csv")

# Contar cuántos elementos hay por cada familia
familia = df['Family'].str.split(',', expand=True).stack()
familia_counts = familia.value_counts()

# Crear el gráfico de pastel
plt.figure(figsize=(18, 18))  # Tamaño máximo del gráfico

# Crear el gráfico de pastel
plt.pie(familia_counts, autopct='%1.1f%%', startangle=90,
        colors=sns.color_palette("pastel"), pctdistance=0.9,  # Mover los porcentajes más lejos
        wedgeprops={'edgecolor': 'black'},
        textprops={'fontsize': 14,'fontweight': 'bold'})  # Aumentar tamaño de los textos (porcentajes)

# Título con tamaño más adecuado y alineado a la izquierda
plt.title('Distribución de Familias de Genes', fontsize=22, loc='left', pad=40)

# Ajustar el gráfico para que sea más legible
plt.axis('equal')  # Para que el gráfico sea circular

# Ajuste de márgenes para maximizar el tamaño del gráfico
plt.subplots_adjust(left=0.05, right=0.85, top=0.95, bottom=0.05)

# Agregar leyenda fuera del gráfico para evitar superposición
plt.legend(familia_counts.index, loc='center left', bbox_to_anchor=(1, 0.5), fontsize=14, title="Familias", title_fontsize=16)

# Mostrar el gráfico
plt.show()
```

```python
# Cargar los datos directamente desde la URL
df = pd.read_csv("https://docs.google.com/spreadsheets/d/1AVefl9QtHKURb3vFBHtjgDNojQPUTClm/export?format=csv")
# Separar los antibióticos por coma y luego contar las ocurrencias de cada uno
antibiotic_group = df['Antibiotic_Group'].str.split(',', expand=True).stack()

# Contar cuántos antibióticos individuales hay
antibioticos_group_counts = antibiotic_group.value_counts()

# Crear el gráfico de barras
plt.figure(figsize=(18, 8))  # Tamaño del gráfico

# Crear el gráfico de barras
ax = antibioticos_group_counts.plot(kind='bar', color="lightsalmon")

# Título y etiquetas
plt.title('Frecuencia de Antibióticos Asociados a ARGs', fontsize=22, loc='left', pad=40)
plt.xlabel('Antibióticos', fontsize=16)
plt.ylabel('Frecuencia', fontsize=16)

# Aumentar el tamaño de los ticks
plt.xticks(rotation=45, ha='right', fontsize=14)
plt.yticks(fontsize=14)

# Añadir los valores encima de las barras
for i, v in enumerate(antibioticos_group_counts):
    ax.text(i, v + 0.2, str(v), ha='center', fontsize=14)

# Ajuste de márgenes
plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.15)

# Mostrar el gráfico
plt.show()
```
